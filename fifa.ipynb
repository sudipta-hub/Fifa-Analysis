```python
import pandas as pd                   ##import nesesaary libraies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
fifa=pd.read_csv("data.csv")
fifa
```

```python
fifa.head()
```


top10_nation_count=fifa.groupby("Nationality").sze()_values(asending=false)



```python
top10_nation_count=fifa.groupby("Nationality").size().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
ax=top10_nation_count.plot.bar(rot=60)
plt.xlabel("Countries")
plt.ylabel("Number of players")
plt.title("Top 10 countries based on players count")
ax.set_ylim(0,1800)

for i in ax.patches:
    ax.text(i.get_x(),i.get_height()+20,str(i.get_height()),fontsize=12)
    
```

```python
fifa.sort_values(by="Overall",ascending=False).head(10)[["Name","Nationality","Club","Overall"]]
```

```python
top10_age_count=fifa.groupby("Age").size().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,8))
ax=top10_age_count.plot.bar(rot=60)
plt.xlabel("Age")
plt.ylabel("Number of players")
plt.title("Graph plot based on age")
ax.set_ylim(0,2000)
 
for i in ax.patches:
    ax.text(i.get_x(),i.get_height()+20,str(i.get_height()),fontsize=12) 

```

```python
top10_club_count=fifa.groupby("Club").size().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,7))
ax=top10_club_count.plot.bar(rot=60)
plt.xlabel("Club")
plt.ylabel("Age")
plt.title("distribution of age based on clubs")
ax.set_ylim(0,70)

for i in ax.patches:
    ax.text(i.get_x(),i.get_height()+20,str(i.get_height()),fontsize=15)
    
```

```python

top10_marking_count=fifa.groupby("Overall").size().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,9))
ax=top10_marking_count.plot.bar(rot=60)
plt.xlabel("Overall")
plt.ylabel("Marking")
plt.title("overall performance based on marking")
ax.set_ylim(0,1800)

for i in ax.patches:
    ax.text(i.get_x(),i.get_height()+20 ,str(i.get_height()),fontsize=11)
```
